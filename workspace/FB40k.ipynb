```python
from knowledge import FBManager
from tripletembed.model import TripletEmbedding
import numpy as np
import pandas as pd

```

```python
fbm = FBManager("data/FB15k/freebase_mtr100_mte100-train.txt", "data/FB15k/freebase_mtr100_mte100-test.txt")
```

```python
s_train, o_train, p_train = fbm.get_train_data()
```

```python
n_entity = fbm.n_inst
n_relation = fbm.n_prop
```

```python
model = TripletEmbedding(n_entity, n_relation, 30, 15, [30], sigma=10e-6, learning_rate=10e-3)
```

```python
model.fit(s_train, o_train, p_train)
```

```python
s_train_negative = []
o_train_negative = []
p_train_negtive = []
n_negative = 5
for s, o, p in zip(s_train, o_train, p_train):
    s_train_negative += [s] * n_negative
    o_train_negative += [o] * n_negative
    nn = 0
    while nn < n_negative:
        p_n = np.random.randint(0, n_relation)
        if p_n != p:
            nn += 1
            p_train_negtive.append(p_n)
```

```python
s_train_pn = np.append(s_train, s_train_negative)
```

```python
o_train_pn = np.append(o_train, o_train_negative)
p_train_pn = np.append(p_train, p_train_negtive)
```

```python
label_train = np.array([1] * s_train.size + [-1] * len(s_train_negative))
```

```python
model.fit(s_train_pn, o_train_pn, p_train_pn, label_train)
```
